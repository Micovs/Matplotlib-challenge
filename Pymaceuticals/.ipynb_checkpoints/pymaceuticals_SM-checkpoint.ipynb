## Observations and Insights 

* Out of the ten drugs, most successful were the drugs Capomulin and Ramicane, with mean tumor volume of 40.67 for the drug, Capomulin and 40.21 for Ramicane, while both drugs had the similar number of mice complete the study.

* It is interesting to note that according to the mean and median tumor volume, Placebo performed better than few of the drug regiments, like Ketapril, Naftisol, and Stelasyn. This finding further questions the effectiveness of the mentioned drugs.

* There were almost identical proportions of male and female mice. The diference is not significant to determin if the sex of the mice corelates with the outcome of the drug.

* Out of the four regimens of interest, only Infubinol had an outlier. 

* There is a strong correlation between mouse weight and average tumor volume, this schould be considered when analyzing the performance of the drugs based of the final tumor volume. To be comperable, all of the drug regiment schould have received mice sample with eaqual weight distribution. 


```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
from scipy.stats import linregress

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
combined_data = pd.merge(mouse_metadata,study_results, on="Mouse ID")

# Display the data table for preview
combined_data.head(20)
```

```python
# Checking the number counts of each Mouse ID
combined_data["Mouse ID"].value_counts().head(10)
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
combined_data_mice_index = combined_data.set_index('Mouse ID')
duplicate_mouse = combined_data_mice_index.loc['g989', :]
duplicate_mouse.head(15)
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
cleaned_data = combined_data_mice_index.drop(index = 'g989')
cleaned_data = cleaned_data.reset_index()
cleaned_data.head(10)
```

```python
# Checking the number counts of each Mouse ID in the clean DataFrame.
cleaned_data["Mouse ID"].value_counts().head(10)
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen


cleaned_data_grouped = cleaned_data.groupby(["Drug Regimen"])
mean = cleaned_data_grouped["Tumor Volume (mm3)"].mean()
median = cleaned_data_grouped["Tumor Volume (mm3)"].median()
var = cleaned_data_grouped["Tumor Volume (mm3)"].var()
std = cleaned_data_grouped["Tumor Volume (mm3)"].std()
sem = cleaned_data_grouped["Tumor Volume (mm3)"].sem()
Summary_statistics_df = pd.DataFrame({"Mean Tumor Volume" :mean, 
                                     "Median Tumor Volume" :median,
                                     "Tumor Volume Variance" :var,
                                     "Tumor Volume Std. Dev." :std,
                                     "Tumor Volume Std. Err." :sem})
Summary_statistics_df
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
cleaned_data_grouped_byID = cleaned_data.groupby(["Mouse ID"])
max_timepoint_df_indexID = pd.DataFrame(cleaned_data_grouped_byID.max()) 
max_timepoint_df = max_timepoint_df_indexID.reset_index()

# Mergeing this group df with the original dataframe to get the tumor volume at the last timepoint
final_tumor_volume_df = pd.merge(max_timepoint_df,cleaned_data, on=("Mouse ID","Timepoint"))

# Grabbing just the needed columns and renaming
final_tumor_volume_df = final_tumor_volume_df[["Mouse ID","Drug Regimen_x","Tumor Volume (mm3)_y"]]
final_tumor_volume_df = final_tumor_volume_df.rename(columns={"Tumor Volume (mm3)_y":"Tumor Volume (mm3)", "Drug Regimen_x":"Drug Regimen"})
final_tumor_volume_df
```

```python
# Creating a list of four treatments and a empty list that will be populated with data for the box plot 
drug_list = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
box_list = []   

# Creating a Df containtng onli the Drug Regimens from the drug list
four_drugs = final_tumor_volume_df[final_tumor_volume_df['Drug Regimen'].isin(drug_list)]
four_drugs = four_drugs.groupby(["Drug Regimen","Mouse ID"])
four_drugs_df=pd.DataFrame(four_drugs.max().round(3))
four_drugs_df
# Reshape dataframe with Drug Regimens as columns by using unstack, also move Last_tumor_size to index using stack
four_drugs_df = four_drugs_df.stack(level=0).unstack(level=0)
four_drugs_df


# Calculate the IQR and quantitatively determine if there are any potential outliers. 

# Quartile calculations for each drug in the drug list
for drug in drug_list:
    quartilles = four_drugs_df[drug].quantile([.25,.5,.75]).round(3)
    lowerq = quartilles[0.25].round(2)
    upperq = quartilles[0.75].round(2)
    iqr = round(quartilles[0.75] - quartilles[0.25],2).round(2)
    lower_bound = round(quartilles[0.25] - (1.5*iqr),2)
    upper_bound = round(quartilles[0.75] + (1.5*iqr),2)
    
    # Droping the NA values and creating a list of lists from the data for each drug name, 
    # That will be used for the box plot    
    box_list.append(list(four_drugs_df[drug].dropna()))
    
    print(f"{drug} IQR data is:")
    print("----------------------")
    print(f"The lower quartile of {drug} is: {lowerq} and the upper quartile is: {upperq}")
    print(f"The interquartile range is: {iqr}")
    print(f"{drug}'s potential outliers are the values below {lower_bound} and the values above {upper_bound}")
    print("--------------------------------------------------------------------")
    
    
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest

# Plot the list of lists using a boxplot utilizing the same y-axis for all plots
fig1, ax = plt.subplots(figsize=(7,5))
ax.set_title("Final tumor volume of each mouse across four regimens of interest", size = 11.55)
ax.set_ylabel("Final Tumor Vol (mm3)")
ax.boxplot(box_list, sym = "r",)
plt.xticks([1,2,3,4],drug_list)

# Save and show the box plot grap
plt.savefig("../Images/box_plot_final_tumor_Vol.png")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

corr = st.pearsonr(cleaned_data_capomulin_x_df["Weight (g)"], cleaned_data_capomulin_x_df["Tumor Volume (mm3)"])
print(f"The correlation between mouse weight and the average tumor volume is {corr[0].round(2)}")
```

```python
# Adding the linear regression equation, line to plot and printing out the r-squared value along with the plot
x_values = cleaned_data_capomulin_x_df["Weight (g)"]
y_values = cleaned_data_capomulin_x_df["Tumor Volume (mm3)"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.annotate(line_eq,(15,35),fontsize=15,color="red")
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")

# Adding x labels and y labels and title
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")

print(f"The r-squared is: {rvalue**2}")

# Save and show the scatter plot graph
plt.savefig("../Images/scater_plot_weight_vs_Avg_Tumor_Vol_regre.png")
plt.show()

```
